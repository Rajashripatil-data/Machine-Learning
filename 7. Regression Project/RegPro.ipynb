## Step 4- Preprocessing for feature selection
cat-ordinal encoding 

```python
X.dtypes
```

```python
cat=list(X.columns[X.dtypes=="object"])
```

```python
cat
```

```python
con=list(X.columns[X.dtypes!="object"])
```

```python
con
```

```python
from sklearn.pipeline import make_pipeline 
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler,OrdinalEncoder
from sklearn.compose import ColumnTransformer
```

```python
num_pipe1= make_pipeline(
    SimpleImputer(strategy="mean"),
    StandardScaler()
)
```

```python
cat_pipe1= make_pipeline(
    SimpleImputer(strategy="constant",fill_value="Not_Avail"),
    OrdinalEncoder()
)
```

```python
pre1= ColumnTransformer(
    [
        ("num",num_pipe1,con),
        ("cat",cat_pipe1, cat)
    ]
).set_output(transform="pandas")
```

```python
X_pre=pre1.fit_transform(X)
X_pre.head()
```

```python
pre1
```

## Apply Feature Selection

```python
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import SequentialFeatureSelector
```

```python
model=LinearRegression()
sel=SequentialFeatureSelector(model,n_features_to_select="auto",direction="backward",n_jobs=-1)
sel.fit(X_pre,Y)
sel_cols=sel.get_feature_names_out()
sel_cols
```

```python
len(sel_cols)
```

```python
sel_cols[0]
```

```python
sel_cols[0].split("__")
```

```python
sel_cols[0].split("__")[1]
```

```python
imp_cols=[col.split("__")[1] for col in sel_cols]
imp_cols
```

```python
X_sel= X[imp_cols]
X_sel
```

## Feature selection is completed

## Step 5- Create a final preprocessing pipeline for X_sel
cat- OneHotEncoder

```python
cat_sel= list(X_sel.columns[X_sel.dtypes=="object"])
con_sel= list(X_sel.columns[X_sel.dtypes!="object"])
```

```python
cat_sel
```

```python
con_sel
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
num_pipe2=make_pipeline(
    SimpleImputer(strategy="median"),
    StandardScaler()
)
```

```python
cat_pipe2= make_pipeline(
    SimpleImputer(strategy="constant",fill_value="Not_Avail"),
    OneHotEncoder(handle_unknown="ignore",sparse_output=False)
)
```

```python
pre2= ColumnTransformer(
    [
        ("num",num_pipe2, con_sel),
        ("cat",cat_pipe2, cat_sel)
    ]
).set_output(transform="pandas")
```

```python
X_sel_pre=pre2.fit_transform(X_sel)
X_sel_pre
```

## Step 6- Train test split

```python
from sklearn.model_selection import train_test_split

xtrain,xtest,ytrain,ytest=train_test_split(X_sel_pre,Y,test_size=0.2,random_state=21)
```

## Step 7- Evaluate all models

```python
from sklearn.metrics import(
    root_mean_squared_error,
    mean_absolute_error,
    mean_absolute_percentage_error,
    r2_score
)
```

```python
def evaluate_model(model,x,y):
    ypred= model.predict(x)
    rmse = root_mean_squared_error(y, ypred)
    mae = mean_absolute_error(y, ypred)
    mape = mean_absolute_percentage_error(y, ypred)
    r2 = r2_score(y, ypred)
    print(f"RMSE : {rmse:.2f}")
    print(f"MAE : {mae:.2f}")
    print(f"MAPE : {mape:.2%}")
    print(f"R2 : {r2:.2%}")
```

```python
print("Linear Regresion Test results :")
evaluate_model(model1, xtest, ytest)
```

```python
print("Ridge Results : ")
evaluate_model(best_ridge, xtest, ytest)
```

```python
print("Lasso Model test results :")
evaluate_model(best_lasso, xtest, ytest)
```
